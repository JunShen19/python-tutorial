### Array & Linked List

```python
my_array = [1, 2, 3, 4, 5]
```

```python
print(my_array[0])  
print(my_array[2])  
```

```python
my_array[1] = 10  
print(my_array)  
```

```python
length = len(my_array)
print(length)
```

```python
my_array.append(6)  
print(my_array)  
```

```python
my_array.insert(2, 7)  
print(my_array)  
```

```python
del my_array[3]  
print(my_array) 
```

```python
my_array.remove(10) 
print(my_array) 
```

```python
slice_array = my_array[1:4]
print(slice_array) 
```
